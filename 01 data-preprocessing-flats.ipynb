```python
def treat_price(x):
    if isinstance(x, float):  # already numeric
        return x
    if isinstance(x, list) and len(x) == 2:
        value, unit = x
        if unit == 'Lac':
            return round(float(value) / 100, 2)
        else:
            return round(float(value), 2)
    elif isinstance(x, list) and len(x) == 1:
        try:
            return round(float(x[0]), 2)
        except:
            return None
    return None  # for unexpected cases

```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head(5)
```

```python
# price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
import re

def clean_price_per_sqft(x):
    if pd.isnull(x):
        return None
    # Remove non-digit, non-dot, non-comma, non-slash, non-₹ characters
    x = re.sub(r'[^\d₹,./]', '', x)
    x = x.split('/')[0]
    x = x.replace('₹', '').replace(',', '').strip()
    try:
        return float(x)
    except:
        return None

```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price_per_sqft)
```
